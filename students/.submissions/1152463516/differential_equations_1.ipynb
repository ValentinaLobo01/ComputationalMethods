```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return y * x ** 2 - 1.2 * y
```

```python
def euler_int(func, intrange, h, yinit):
    x = np.arange(intrange[0], intrange[-1] + h, h)
    y = [yinit]
    for i in range(len(x) - 1):
        y.append(y[i] + func(x[i], y[i]) * h)
    return x, y
```

```python
fig,ax=plt.subplots(1,1)

intrange=[-2.0,2.0]
h=0.5
yinit=0.76
x1,y1=euler_int(f,intrange,h,yinit)
ax.plot(x1,y1,'-r',label='h=0.5')

intrange=[-2.0,2.0]
h=0.25
yinit=0.76
x2,y2=euler_int(f,intrange,h,yinit)
ax.plot(x2,y2,'-b',label='h=0.25')

intrange=[-2.0,2.0]
h=0.05
yinit=0.76
x3,y3=euler_int(f,intrange,h,yinit)
ax.plot(x3,y3,'-g',label='h=0.05')

xexact=np.linspace(intrange[0],intrange[-1],101)
yexact=np.exp((xexact**3)/3-1.2*xexact)
ax.plot(xexact,yexact,'-k',label='exact')

ax.set_xlabel('x')
ax.set_ylabel('ODE')
ax.legend()
plt.show()
```
