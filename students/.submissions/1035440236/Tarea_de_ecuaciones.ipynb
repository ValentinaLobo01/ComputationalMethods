```python
# Ejercicios ecuaciones diferenciales:
# Segunda clase:
import matplotlib.pyplot as plt    # Importamos la libreria
import numpy as np  # Importamos numpy
from scipy import integrate   # Importante importar scipy integrate
# Importamos las librerias.
import numpy as np
from scipy.integrate import odeint 
from scipy.integrate import solve_ivp
```

```python
# Implementacion con odeint.
# Tenemos lo siguiente.# Ejercicio:

def odes1(x, t):  # Definimos nuestra funcion
  sigma=10 # Valor de sigma
  rho=28  # Valor de rho
  beta=8/3  # Valor de bheta.
  x1=x[0]  # x
  y1=x[1]  # y
  z1=x[2]  # z
  dxdt=sigma*(y1-x1)
  dydt=x1*(sigma-z1)-y1
  dzdt=x1*y1-beta*z1
  return [dxdt, dydt, dzdt]  # Retornamos la funcion.

x01 = [1, 1, 1]

t = np.linspace(0,15,1000)  # Intervalo t.
x = odeint(odes1,x01,t)
x1 = x[:,0]  # Columna cero
y1 = x[:,1]  # Columna 1
z1 = x[:,2]   # Columna 2.

plt.plot(t,x1)
plt.plot(t,y1)   # Grafica
plt.plot(t,z1)
plt.grid()
plt.xlabel("eje t")
plt.ylabel("y")
plt.title("Grafica")
plt.show()
```
