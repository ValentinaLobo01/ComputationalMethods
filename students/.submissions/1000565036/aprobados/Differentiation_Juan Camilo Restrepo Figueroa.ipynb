```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

```python
def myfunc(x,a):
    return a*np.sin(x)
```

Gráfica de la función:

```python
x=np.linspace(0,2*np.pi,50)

plt.figure(figsize=(12,8))
plt.plot(x,np.sin(x), color="black", label="Si a=1")
plt.legend()
plt.plot(x,2*np.sin(x), color="blue", label="Si a=2")
plt.legend()
plt.plot(x,3*np.sin(x), color="Red", label="Si a=3")
plt.legend()
plt.grid(True)
plt.show
```

Derivada de la función en $x=\frac{\pi}{2}$:

```python
d=derivative(myfunc,np.pi*(1/2),args=(2,)) 
d
```

Definamos la derivada analítica:

```python
def myfuncderiv(x,a):
    return a*np.cos(x)
```

Grafiquemos `myfunc`, `myfuncderiv` y la derivada en $x=\frac{\pi}{2}$ para a=1:

```python
x=np.linspace(0,2*np.pi,50)

plt.figure(figsize=(12,8))
plt.plot(x,np.sin(x), color="black", label="MyFunc")
plt.legend()
plt.plot(x,np.cos(x), color="red", label="MyFuncDeriv")
plt.legend()
plt.plot(np.pi*(1/2),d,"o", label="Derivada")
plt.legend()
plt.grid(True)
plt.show
```
