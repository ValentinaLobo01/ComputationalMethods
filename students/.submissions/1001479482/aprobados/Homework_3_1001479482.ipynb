```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,a):
  return a*np.sin(x)
```

```python
#function
x = np.linspace(0,2*np.pi, 100)
f_x = f(x,1)
```

```python
#Analytical derivate
def fprime(x,a):
  return a*np.cos(x)

fprime_x = fprime(x,1)
```

```python
#misc derivate
misc = derivative(f, x, dx=.1, n=1, args=(1,))
```

```python
#Plot and comparisson
plt.figure(figsize=(10,6))
plt.plot(x, f_x,label='$f(x)=sin(x)$')
plt.plot(x, fprime_x,label='$Analytical \enspace derivate$')
plt.plot(x, misc,label='$Scipy \enspace derivate$')
plt.legend()
plt.plot()
```

As the plot is showing, the difference between the analytical and the scipy derivative is very small. We cannot see the analytical derivate because scipy derivative is above it. This difference depends on the value that we assign to $dx$ in the parameters of scipy derivative.

```python
differences=abs(misc-fprime_x)
differences
```

```python
np.mean(differences)
```

```python
max(differences)
```

```python
min(differences)
```

**Finally, we can see that the difference between the two methods is almost despicable, where the maximum value equal to 0.002.**
