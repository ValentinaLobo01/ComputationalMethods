## Homework:  Boundary Value Problems


```python
#Libraries
from scipy.integrate import solve_bvp
import numpy as np
import matplotlib.pyplot as plt

#functions 
def derivs(x,y):
    dydx=np.zeros((y.shape))
    dydx[0]=y[1]
    dydx[1]= 0.01 * y[0] - 0.2
    return dydx

def bc(ya,yb):
    residuals=np.array([0.0,0.0])
    residuals[0]=ya[0] - 40
    residuals[1]= yb[0] -200
    print("z(0) = {}".format(ya[1]))
    return residuals


x_steps = 100
x = np.linspace(0,10,x_steps)

y0 = np.ones((2,x_steps))

sol = solve_bvp(derivs, bc, x, y0)

#Plot
plt.figure(figsize=(8, 5), dpi=80)
plt.scatter(0, 40, color ="red")
plt.scatter(10, 200, color ="red")
plt.plot(sol.x, sol.y[0])
plt.xlabel("x")
plt.ylabel("T(x)")
plt.grid();
```
