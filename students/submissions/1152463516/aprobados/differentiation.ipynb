```python
import numpy as np

def myfunc(x, a):
    return a * np.sin(x)

def d_myfunc(x, a):
    return a * np.cos(x)
```

```python
import numpy as np
import scipy.misc as misc
import matplotlib.pyplot as plt

a = 1
xn = np.linspace(0, 10, 100)
dyn = d_myfunc(xn, a)
scipy_dyn = misc.derivative(myfunc, xn, args=(a, ))

plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(10, 5))
fig.tight_layout()

ax.plot(xn, dyn, label='analytical derivative')
ax.plot(xn, scipy_dyn, ls='--', label='scipy derivative')
ax.legend()

ax.set_title('acos(x)')
```
