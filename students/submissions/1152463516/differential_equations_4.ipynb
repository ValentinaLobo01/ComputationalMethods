```python
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
```

```python
def derivatives(x, y):
    return y[1], 0.01 * y[0] - 0.2

def bc(ya, yb):
    return ya[0] - 40, yb[0] - 200
```

```python
# define the array of independent variable with the interval of the conditions
t = np.linspace(0, 10, 20)

# define the array of dependent variables, one row for each DE
y = np.zeros((2, t.size))

# array with one row filled with values of T, and one with z
sol = integrate.solve_bvp(derivatives, bc, t, y).y

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(t, sol[0], marker='o')
ax.set_ylabel('T(x)')
ax.set_xlabel('x')
```
