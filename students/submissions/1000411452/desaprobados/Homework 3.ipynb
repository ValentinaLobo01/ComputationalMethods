```python
import numpy as np
import scipy
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
def myfunc(x,a = 1):
    return a*np.sin(x)
```

```python
x = np.linspace(0,2*np.pi,20)
y = myfunc(x,a=1.0)

plt.plot(x,y)
```

```python
#Derivative
d = derivative(myfunc, 1.0, dx=1.0, n=1, args=(), order=3) #In one point
d
```
