Use the scipy.misc.derivative package find the derivative of the following function:


def myfunc(x,a):
 
 return a*np.sin(x)


where 'a' is an arbitrary variable passed to the scipy routine

*hint: read the documentation of the scipy.misc.derivative and check how one can pass arguments to the fuction

Plot the function, the derivative and compare with analytical results of the function for a=1

```python
#Libraries
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
#Function
def myfunc(x,a):
    return a*np.sin(x)

# AnalyticalDerivative

def fprime(x, a):
    return a* np.cos(x)
```

```python
dx = 0.01
x0 = 2
a = 1

print("function result : {} \n".format(myfunc(x0, a)))
print("analytical result: {} \n ".format(fprime(x0, a)))
print("result with sp.misc.derivative: {} ".format(derivative(myfunc ,x0, dx , args=[a])))
```

```python
x = np.linspace(-2*np.pi , 2*np.pi, 100)

#Plot derivates 
x2 = np.linspace((x0-3), (x0+3), 100)
y = derivative(myfunc,x0,dx=dx, args = [a])*(x2 - x0) + myfunc(x0,a)
y2 = fprime(x0, a) * (x2 -x0) + myfunc(x0, a)

plt.figure(figsize=(10, 8))
plt.plot(x, myfunc(x, a), color ="black", label = "Function ")
plt.plot(x2, y, label = "Scipy.misc.derivate()")
plt.plot(x2, y2, label = "Analytical derivate",linestyle = "--")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid();
```
